## 1.Load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction import DictVectorizer
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-03-churn-prediction/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

## 2.EDA

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df.totalcharges = pd.to_numeric(df.totalcharges, errors='coerce')
df.totalcharges = df.totalcharges.fillna(0)
```

```python
df.isnull().sum()
```

```python
categorical = list(df.dtypes[df.dtypes == 'object'].index)
numerical   =list(df.dtypes[df.dtypes != 'object'].index)
```

```python
for col in categorical:
    df[col] = df[col].str.lower().str.replace(' ','_')
    
```

```python
categorical.remove('customerid')
categorical.remove('churn')
```

```python
df[categorical].describe()
```

```python
df['churn'] = df['churn'] == 'yes'
```

```python
df.churn= df['churn'].astype(int)
```

## ROC 

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
scores = []
thresholds = np.linspace(0, 1, 10001)

for t in thresholds:
    actual_positive = (y_val == 1)
    actual_negative = (y_val == 0)
    
    predict_positive = (y_pred >= t)
    predict_negative = (y_pred < t)

    tp = (predict_positive & actual_positive).sum()
    tn = (predict_negative & actual_negative).sum()

    fp = (predict_positive & actual_negative).sum()
    fn = (predict_negative & actual_positive).sum()
    
    scores.append((t, tp, fp, fn, tn))
```

```python
columns = ['threshold', 'tp', 'fp', 'fn', 'tn']
df_scores = pd.DataFrame(scores, columns=columns)

df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)
```

```python
fpr_m ,tpr_m, thresholds_m = roc_curve(y_val, y_pred)
```

## K-Fold validation

```python
def train(train_set, y_train, C=1):
    dicts_train = train_set[categorical +numerical].to_dict(orient = 'records')
    dv = DictVectorizer(sparse=False)
    X_train = dv.fit_transform(dicts_train)

    model = LogisticRegression(C=C,max_iter= 1000)
    model.fit(X_train, y_train)
    return model, dv
```

```python
def predict(val_set, dv, model):
    val_dicts = val_set[categorical +numerical].to_dict(orient = 'records')
    X_val = dv.transform(val_dicts)

    y_pred = model.predict_proba(X_val)[:, 1]
    return y_pred
```

```python
from sklearn.model_selection import KFold
from tqdm.auto import tqdm
```

```python
n_splits = 5

for C in tqdm([0.001, 0.01, 0.1, 0.5, 1, 5, 10]):
    kfold = KFold(n_splits=n_splits, shuffle=True, random_state=1)
    scores = []
    for train_idx,val_idx in tqdm(kfold.split(full_train)):
        train_set = full_train.iloc[train_idx]
        val_set   = full_train.iloc[val_idx]
    
        y_train   = train_set.churn.values
        y_val     = val_set.churn.values
    
        model, dv = train(train_set, y_train,C= C)
        y_pred    = predict(val_set,dv,model)
        auc       = roc_auc_score(y_val,y_pred)
    
        scores.append(auc)
    print('C=%s %.3f +- %.3f' % (C, np.mean(scores), np.std(scores)))

print(scores)
```
